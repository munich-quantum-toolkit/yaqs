```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import pickle
import os

# ----- Configure global style -----
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "axes.labelsize": 14,
    "font.size": 13,
    "axes.titlesize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "axes.linewidth": 1.0,
    "legend.fontsize": 12,
    "figure.figsize": (12, 6)
})

L = 32  # Number of sites/bonds
num_files = 4
dt = 0.1

data_list = []
nsteps = None

# Step 1: Load all data and determine global min/max
vmin = float('inf')
vmax = float('-inf')
# gammas = [1e-3, 1e-1, 1e1, 1e2, 1e3]
gammas = [1e-3, 1e-2, 1e-1, 1e2]

for idx in range(4):
    filename = f"32/correlator_{idx}.pickle"
    if not os.path.exists(filename):
        print(f"File {filename} not found, skipping.")
        data_list.append(None)
        continue

    with open(filename, 'rb') as f:
        observables = pickle.load(f)

    data = []
    for observable in observables[0]:
        if observable.gate.name not in ["runtime_cost", "max_bond", "total_bond"]:
            data.append(observable.results)
    # data = [observable.results for observable in observables[0]]  # shape: [timesteps, sites]
    data = np.array(data).T
    data = np.flip(data)  # Rotate 180Â°

    vmin = min(vmin, np.min(data))
    vmax = max(vmax, np.max(data))
    data_list.append(data)

    if nsteps is None:
        nsteps = data.shape[1]

# Step 2: Plot with shared color scale
fig, axes = plt.subplots(1, num_files, sharey=True)
# Set up the symmetric log normalization
norm = mcolors.SymLogNorm(
    linthresh=1e-2,  # linear region within [-1e-2, 1e-2]
    linscale=1.0,
    vmin=0,
    vmax=vmax,
    base=10
)

for idx, ax in enumerate(axes):
    data = data_list[idx]
    if data is None:
        continue

    im = ax.imshow(
        data,
        aspect="auto",
        extent=(0, L, 50, 0),
        cmap='viridis',
        # norm=norm,
        vmin=0,
        vmax=vmax
    )
    ax.set_title(f"$\\gamma=$ {gammas[idx]}")
    ax.set_xlabel("Bond")
    if idx == 0:
        ax.set_label("Time")

# Step 3: Shared colorbar
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.8)
cbar.set_label("Correlation")

plt.savefig("dynamics.pdf", dpi=300)
# plt.tight_layout()
plt.show()
```
